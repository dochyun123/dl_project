```python
def update_df_with_item(df_source, df_target, item_name, target_column_name):
    """
    Update df_target DataFrame with values from df_source based on a specific item_name.
    
    Parameters:
    df_source (DataFrame): Source DataFrame containing the data.
    df_target (DataFrame): Target DataFrame to be updated.
    item_name (str): Name of the item to filter on in df_source.
    target_column_name (str): Name of the column in df_target where the data will be updated.
    """
    for index, row in df_target.iterrows():
        # Find matching value in df_source
        value = df_source[(df_source['회사명'] == row['회사명']) & (df_source['항목명'] == item_name)]['당기 1분기']
        
        # Assign the value to the target column in df_target
        if not value.empty:
            df_target.at[index, target_column_name] = value.iloc[0]
```

```python
#Example usage
update_df_with_item(BS_data, df_2022_BS, '유동자산', '당기_유동자산')
update_df_with_item(BS_data, df_2022_BS, '비유동자산', '당기_비유동자산')
update_df_with_item(BS_data, df_2022_BS, '유동부채', '당기_유동부채')
update_df_with_item(BS_data, df_2022_BS, '비유동부채', '당기_비유동부채')
update_df_with_item(BS_data, df_2022_BS, '자본총계', '당기_자본')
update_df_with_item(BS_data, df_2022_BS, '부채및자본총계', '당기_부채 및 자본총계')
update_df_with_item(BS_data, df_2022_BS, '부채총계', '당기_부채총계')
update_df_with_item(BS_data, df_2022_BS, '자산총계', '당기_자산총계')
#Add more lines as needed for other items like '자본', etc.
```
