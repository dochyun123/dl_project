```python
df = pd.read_excel("2022_data/2022_consolidated/2022_CF.xlsx")

# 예시 DataFrame을 생성하거나 불러옵니다.
# df = pd.read_csv('your_dataframe.csv') # 데이터프레임 불러오기

# '영업활동'을 포함하는 행 필터링
df_operating = df[df['항목명'].str.contains('유동자산')]

# '투자활동'을 포함하는 행 필터링
df_investing = df[df['항목명'].str.contains('투자활동')]

# '재무활동'을 포함하는 행 필터링
df_financing = df[df['항목명'].str.contains('재무활동')]

# 결과를 확인합니다.
print(df_operating[['종목코드', '회사명', '당기', '전기', '전전기']])
print(df_investing[['종목코드', '회사명', '당기', '전기', '전전기']])
print(df_financing[['종목코드', '회사명', '당기', '전기', '전전기']])
```
