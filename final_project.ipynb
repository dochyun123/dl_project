```python
corp_list = dart.get_corp_list()  # all corps : 103463 corps in list
start_date = "20200101"
end_date = "20201231"


def extract_fs(index, begin = '20200101'):
    corp_list = dart.get_corp_list()  # corp_code, corp_name, stock_code = None
    corp_name = corp_list[index].corp_name
    corporation = corp_list.find_by_corp_name(corp_name, exactly=True)[0]
    origin = corporation.extract_fs(
        bgn_de=begin,
        end_de=None,
        fs_tp=("bs", "is", "cis", "cf"),
        separate=True,
        report_tp="annual",
        lang="ko",
        separator=True,
        dataset="xbrl",
        cumulative=False,
        progressbar=True,
        skip_error=True,
        last_report_only=True,
    )
    bs_fs = origin["bs"]
    is_fs = origin["is"]
    cis_fs = origin["cis"]
    cf_fs = origin["cf"]
    return bs_fs, is_fs, cis_fs, cf_fs
```
